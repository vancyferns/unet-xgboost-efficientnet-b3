```python
import cv2, numpy as np
from skimage import img_as_float
from albumentations import Compose, Resize, Normalize
from albumentations.pytorch import ToTensorV2

def preprocess_raw_image_cv2(img_bgr, out_size=(IMG_SIZE,IMG_SIZE), apply_clahe=True, blur_ksize=(5,5)):
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    if apply_clahe:
        lab = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2LAB)
        l,a,b = cv2.split(lab)
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
        l2 = clahe.apply(l)
        lab = cv2.merge((l2,a,b))
        img_rgb = cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)
    if blur_ksize is not None:
        img_rgb = cv2.GaussianBlur(img_rgb, blur_ksize, 0)
    img_rgb = cv2.resize(img_rgb, out_size, interpolation=cv2.INTER_AREA)
    return img_rgb

print("Preprocess ready.")
```

```python
# === Cell 5 (FIXED): Use U-Net for vessel segmentation using Frangi pseudo-labels ===
# Assumes: preprocess_raw_image_cv2(img_bgr, out_size=(IMG_SIZE,IMG_SIZE)) exists (from Cell 4)
# and IMAGES_DIR, IMG_SIZE, DEVICE are defined.

!pip install -q albumentations==1.3.1

import os, cv2, numpy as np, math, random, tqdm
from skimage.color import rgb2gray
from skimage.filters import frangi
from skimage import img_as_ubyte
import skimage.filters as skf
from tqdm import tqdm
import torch, torch.nn as nn, torch.nn.functional as F, torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from albumentations import Compose, Resize, Normalize, HorizontalFlip, VerticalFlip, ShiftScaleRotate
from albumentations.pytorch import ToTensorV2

# --- hyperparams (tweak if needed) ---
FRANGI_SCALE_RANGE = (1,6)
FRANGI_SCALE_STEP = 0.5
FRANGI_BETA1 = 0.5
FRANGI_BETA2 = 15.0
pseudo_threshold_factor = 0.25  # lower -> more permissive pseudo-mask (0..1 fraction of mean+std)
MASK_DIR = "/content/frangi_pseudo_for_unet"
PROB_DIR = "/content/frangi_probs_for_unet"
os.makedirs(MASK_DIR, exist_ok=True)
os.makedirs(PROB_DIR, exist_ok=True)

# U-Net training params
EPOCHS = 12
BATCH_SIZE = 4
LR = 1e-3
BASE_C = 32
val_frac = 0.15
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED); np.random.seed(RANDOM_SEED); random.seed(RANDOM_SEED)

# --- 1) collect images (only those in IMAGES_DIR matching 100..163 if present) ---
image_files = sorted([os.path.join(IMAGES_DIR,f) for f in os.listdir(IMAGES_DIR) if f.lower().endswith(('.png','.jpg','.jpeg'))])
# try to restrict to numeric 100..163 if filenames include numbers
def numeric_from_name(fname):
    import re
    m = re.search(r'(\d{2,3})', fname)
    return int(m.group(1)) if m else None
selected = [p for p in image_files if numeric_from_name(os.path.basename(p)) and 100 <= numeric_from_name(os.path.basename(p)) <= 163]
if len(selected) == 0:
    selected = image_files  # fallback
print("Training images count:", len(selected))

# --- 2) create Frangi probability maps and permissive binary pseudo-masks ---
print("Computing Frangi probability maps and pseudo-masks (this may take a few minutes)...")
# --- FIX 1: Changed tqdm.tqdm to tqdm ---
for p in tqdm(selected):
    name = os.path.basename(p)
    prob_path = os.path.join(PROB_DIR, name)
    mask_path = os.path.join(MASK_DIR, name)
    if os.path.exists(prob_path) and os.path.exists(mask_path):
        continue
    bgr = cv2.imread(p)
    if bgr is None:
        print("Unreadable:", p); continue
    rgb = preprocess_raw_image_cv2(bgr, out_size=(IMG_SIZE, IMG_SIZE), apply_clahe=True, blur_ksize=(5,5))
    gray = rgb2gray(rgb)
    # compute frangi (try with beta params if supported)
    try:
        v = frangi(gray, scale_range=FRANGI_SCALE_RANGE, scale_step=FRANGI_SCALE_STEP, beta1=FRANGI_BETA1, beta2=FRANGI_BETA2)
    except TypeError:
        v = frangi(gray, scale_range=FRANGI_SCALE_RANGE, scale_step=FRANGI_SCALE_STEP)
    # normalize
    vmin, vmax = float(np.nanmin(v)), float(np.nanmax(v))
    vnorm = (v - vmin) / (vmax - vmin + 1e-8)
    v_ubyte = img_as_ubyte(vnorm)  # 0..255
    cv2.imwrite(prob_path, v_ubyte)
    # permissive threshold: mean + k*std (k = pseudo_threshold_factor * (std multiplier scale))
    # compute threshold in 0..255 space
    mean = v_ubyte.mean()
    std = v_ubyte.std()
    thresh = mean + (pseudo_threshold_factor * std * 3.0)  # multiply to scale sensitivity; tweak if needed
    # fallback to Otsu if you want stricter
    try:
        otsu_t = skf.threshold_otsu(v_ubyte)
        # choose the smaller of otsu and our permissive threshold to keep fine vessels
        thresh = min(thresh, otsu_t)
    except Exception:
        pass
    mask_bin = (v_ubyte >= thresh).astype('uint8') * 255
    cv2.imwrite(mask_path, mask_bin)

print("Saved frangi prob maps ->", PROB_DIR)
print("Saved permissive pseudo-masks ->", MASK_DIR)

# --- 3) Dataset & transforms ---
train_transform = Compose([
    Resize(IMG_SIZE, IMG_SIZE),
    HorizontalFlip(p=0.5),
    VerticalFlip(p=0.2),
    ShiftScaleRotate(shift_limit=0.06, scale_limit=0.12, rotate_limit=20, p=0.6),
    Normalize(mean=0.5, std=0.5),
    ToTensorV2(),
])

valid_transform = Compose([
    Resize(IMG_SIZE, IMG_SIZE),
    Normalize(mean=0.5, std=0.5),
    ToTensorV2(),
])

class VesselDataset(Dataset):
    def __init__(self, image_paths, masks_dir, transform=None):
        self.image_paths = image_paths
        self.masks_dir = masks_dir
        self.transform = transform

    def __len__(self):
        return len(self.image_paths)

    def __getitem__(self, idx):
        ip = self.image_paths[idx]
        name = os.path.basename(ip)
        im = cv2.cvtColor(cv2.imread(ip), cv2.COLOR_BGR2RGB)

        # Note: This resize is redundant since your transform already does it,
        # but it's not the cause of the error.
        im = cv2.resize(im, (IMG_SIZE, IMG_SIZE))

        mask_path = os.path.join(self.masks_dir, name)
        m = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE) if os.path.exists(mask_path) else np.zeros((IMG_SIZE,IMG_SIZE),dtype=np.uint8)

        # Apply transforms (which includes ToTensorV2)
        augmented = self.transform(image=im, mask=m) if self.transform else {"image":im, "mask":m}
        img_t = augmented['image']
        mask_t = augmented['mask']  # This is now a PyTorch Tensor

        # --- START OF FIX ---
        # Original line 1: mask_t = (mask_t > 127).astype('float32')
        # Original line 2: mask_t = torch.from_numpy(mask_t).unsqueeze(0)

        # .astype('float32') becomes .float() for PyTorch tensors
        # (mask_t > 127) creates a boolean tensor
        mask_t = (mask_t > 127).float()

        # torch.from_numpy() is removed because mask_t is already a tensor
        # .unsqueeze(0) adds the required channel dimension (H, W) -> (1, H, W)
        mask_t = mask_t.unsqueeze(0)
        # --- END OF FIX ---

        # img_t and mask_t are already float tensors, no need for .float() again
        return img_t, mask_t

# train/val split
n = len(selected)
nval = max(1, int(n * val_frac))
train_paths = selected[:-nval]
val_paths = selected[-nval:]
train_ds = VesselDataset(train_paths, MASK_DIR, transform=train_transform)
val_ds = VesselDataset(val_paths, MASK_DIR, transform=valid_transform)
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)

print("Train/Val sizes:", len(train_ds), len(val_ds))

# --- 4) define UNet (compact) ---
class DoubleConv(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
            nn.Conv2d(out_ch, out_ch, kernel_size=3, padding=1, bias=False),
            nn.BatchNorm2d(out_ch), nn.ReLU(inplace=True),
        )
    def forward(self, x): return self.net(x)

class Down(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.pool_conv = nn.Sequential(nn.MaxPool2d(2), DoubleConv(in_ch, out_ch))
    def forward(self, x): return self.pool_conv(x)

class Up(nn.Module):
    def __init__(self, in_ch, out_ch, bilinear=True):
        super().__init__()
        if bilinear:
            self.up = nn.Upsample(scale_factor=2, mode='bilinear', align_corners=True)
            self.conv = DoubleConv(in_ch, out_ch)
        else:
            self.up = nn.ConvTranspose2d(in_ch//2, in_ch//2, kernel_size=2, stride=2)
            self.conv = DoubleConv(in_ch, out_ch)
    def forward(self, x1, x2):
        x1 = self.up(x1)
        diffY = x2.size()[2] - x1.size()[2]
        diffX = x2.size()[3] - x1.size()[3]
        x1 = F.pad(x1, [diffX//2, diffX - diffX//2, diffY//2, diffY - diffY//2])
        x = torch.cat([x2, x1], dim=1)
        return self.conv(x)

class UNetSmall(nn.Module):
    def __init__(self, n_channels=3, n_classes=1, base_c=BASE_C):
        super().__init__()
        self.inc = DoubleConv(n_channels, base_c)
        self.down1 = Down(base_c, base_c*2)
        self.down2 = Down(base_c*2, base_c*4)
        self.down3 = Down(base_c*4, base_c*8)
        self.down4 = Down(base_c*8, base_c*8)
        self.up1 = Up(base_c*16, base_c*4)
        self.up2 = Up(base_c*8, base_c*2)
        self.up3 = Up(base_c*4, base_c)
        self.up4 = Up(base_c*2, base_c)
        self.outc = nn.Conv2d(base_c, n_classes, kernel_size=1)
    def forward(self, x):
        x1 = self.inc(x); x2 = self.down1(x1); x3 = self.down2(x2); x4 = self.down3(x3); x5 = self.down4(x4)
        x = self.up1(x5,x4); x = self.up2(x,x3); x = self.up3(x,x2); x = self.up4(x,x1)
        return self.outc(x)

model = UNetSmall(n_channels=3, n_classes=1).to(DEVICE)
print("Model params:", sum(p.numel() for p in model.parameters())/1e6, "M")

# --- 5) loss, optimizer, scheduler ---
bce = nn.BCEWithLogitsLoss()
def dice_loss_logits(pred_logits, target, eps=1e-6):
    pred = torch.sigmoid(pred_logits)
    num = 2 * (pred * target).sum(dim=(2,3)) + eps
    den = (pred + target).sum(dim=(2,3)) + eps
    return (1 - (num/den)).mean()

def combined_loss(logits, targets, w_dice=1.0, w_bce=1.0):
    return w_dice * dice_loss_logits(logits, targets) + w_bce * bce(logits, targets)

opt = optim.Adam(model.parameters(), lr=LR)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(opt, factor=0.5, patience=3)

# --- 6) training loop ---
best_val = 0.0
CKPT = "/content/unet_pseudo_best.pth"
for epoch in range(1, EPOCHS+1):
    model.train()
    train_losses = []
    # --- FIX 2: Changed tqdm.tqdm to tqdm ---
    pbar = tqdm(train_loader, desc=f"Train E{epoch}/{EPOCHS}")
    for imgs, masks in pbar:
        imgs = imgs.to(DEVICE); masks = masks.to(DEVICE)
        opt.zero_grad()
        logits = model(imgs)
        loss = combined_loss(logits, masks)
        loss.backward()
        opt.step()
        train_losses.append(loss.item())
        pbar.set_postfix({'loss': float(loss.item())})
    avg_train = float(np.mean(train_losses)) if train_losses else 0.0

    # validation
    model.eval()
    val_losses = []; val_dices = []
    with torch.no_grad():
        for imgs, masks in val_loader:
            imgs = imgs.to(DEVICE); masks = masks.to(DEVICE)
            logits = model(imgs)
            loss = combined_loss(logits, masks)
            val_losses.append(loss.item())
            preds = (torch.sigmoid(logits) > 0.5).float()
            inter = (preds * masks).sum(dim=(2,3))
            denom = (preds + masks).sum(dim=(2,3))
            dice_batch = ((2*inter + 1e-6)/(denom + 1e-6)).mean().item()
            val_dices.append(dice_batch)
    val_loss = float(np.mean(val_losses)) if val_losses else 0.0
    val_dice = float(np.mean(val_dices)) if val_dices else 0.0
    scheduler.step(val_loss)
    print(f"Epoch {epoch} train_loss={avg_train:.4f} val_loss={val_loss:.4f} val_dice={val_dice:.4f} lr={opt.param_groups[0]['lr']:.6f}")
    if val_dice > best_val:
        best_val = val_dice
        torch.save({'epoch':epoch,'model_state':model.state_dict(),'val_dice':best_val}, CKPT)
        print("Saved best ->", CKPT)

print("Training finished. Best val dice:", best_val)

# --- 7) inference: save refined masks and overlays ---
OUT_REFINED = "/content/unet_refined_masks"
OUT_OVERLAY = "/content/unet_overlays"
os.makedirs(OUT_REFINED, exist_ok=True)
os.makedirs(OUT_OVERLAY, exist_ok=True)

# load best
if os.path.exists(CKPT):
    ck = torch.load(CKPT, map_location=DEVICE)
    model.load_state_dict(ck['model_state'])
    print("Loaded checkpoint epoch", ck.get('epoch'), "val_dice", ck.get('val_dice'))

model.eval()
# --- FIX 3: Changed tqdm.tqdm to tqdm ---
for p in tqdm(selected):
    name = os.path.basename(p)
    orig = cv2.imread(p)
    H,W = orig.shape[:2]
    # prepare input
    inp = preprocess_raw_image_cv2(orig, out_size=(IMG_SIZE, IMG_SIZE), apply_clahe=True, blur_ksize=(5,5))
    inp_t = (inp.astype('float32')/255.0 - 0.5)/0.5
    inp_t = np.transpose(inp_t, (2,0,1))[None,...]
    tensor = torch.from_numpy(inp_t).to(DEVICE).float()
    with torch.no_grad():
        logits = model(tensor)
        prob = torch.sigmoid(logits).squeeze().cpu().numpy()
    mask = (prob > 0.5).astype('uint8') * 255
    mask_big = cv2.resize(mask, (W,H), interpolation=cv2.INTER_NEAREST)
    cv2.imwrite(os.path.join(OUT_REFINED, name), mask_big)
    # overlay: colorize mask red and apply CLAHE to original for better visibility
    img_vis = cv2.cvtColor(orig, cv2.COLOR_BGR2RGB)
    lab = cv2.cvtColor(img_vis, cv2.COLOR_RGB2LAB)
    l,a,b = cv2.split(lab)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    l2 = clahe.apply(l)
    img_clahe = cv2.cvtColor(cv2.merge((l2,a,b)), cv2.COLOR_LAB2RGB)
    red_mask = np.zeros_like(img_clahe); red_mask[...,0] = mask_big
    overlay = cv2.addWeighted(img_clahe, 0.75, red_mask, 0.45, 0)
    # save overlay BGR
    cv2.imwrite(os.path.join(OUT_OVERLAY, name), cv2.cvtColor(overlay, cv2.COLOR_RGB2BGR))

print("Refined masks saved ->", OUT_REFINED)
print("Overlay images saved ->", OUT_OVERLAY)

# === End of cell: U-Net training + inference ===
```

```python
# UPDATED: Re-extract vascular features from UNet masks and train XGBoost (stable + CV + saved outputs)
import os, sys, math, time
import numpy as np
import pandas as pd
import cv2
from tqdm import tqdm
from skimage.morphology import skeletonize, remove_small_objects, opening, disk
from scipy import ndimage as ndi
from skimage.measure import label, regionprops
import xgboost as xgb
from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.metrics import roc_auc_score, accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import StandardScaler

# ---------- CONFIG ----------
POSSIBLE_MASK_DIRS = [
    "/content/unet_refined_masks",
    "/content/unet_refined_mask",
    "/content/unet_refined",
    "/content/UNet_refined_masks",
    "/content/UNet_refined",
    "/content/unet_refined_masks/"
]
MASK_DIR = next((p for p in POSSIBLE_MASK_DIRS if os.path.exists(p) and os.path.isdir(p)), None)
if MASK_DIR is None:
    raise RuntimeError("UNet refined masks folder not found. Set MASK_DIR to your masks folder (looked in common paths).")
print("Using MASK_DIR =", MASK_DIR)

LABELS_CSV = "/content/labels_100_163.csv"   # primary expected labels file
if not os.path.exists(LABELS_CSV):
    alt = "/content/entries_100_to_163.csv"
    if os.path.exists(alt):
        LABELS_CSV = alt
    else:
        LABELS_CSV = None

MIN_OBJECT_SIZE = 50   # set 0 to disable
MORPH_OPEN_RADIUS = 2  # set 0 to disable

OUT_FEATURES = "/content/vascular_features_unet.csv"
OUT_FEATURES_PRED = "/content/vascular_features_unet_with_preds.csv"
OUT_MODEL = "/content/xgb_unet_model.json"
OUT_FI = "/content/xgb_feature_importance.csv"
OUT_REPORT = "/content/xgb_classification_report.txt"

RANDOM_STATE = 42
N_FOLDS = 5  # stratified CV folds (set 3 or 5 depending on data)

# ---------- feature helper functions ----------
def fractal_box_count(img_bin):
    min_side = min(img_bin.shape)
    if min_side < 4:
        return float('nan')
    max_pow = int(math.floor(math.log2(min_side)))
    sizes = 2 ** np.arange(max_pow, 1, -1)
    counts = []
    for S in sizes:
        if S <= 1: continue
        nX = int(np.ceil(img_bin.shape[1] / S))
        nY = int(np.ceil(img_bin.shape[0] / S))
        count = 0
        for i in range(nY):
            for j in range(nX):
                patch = img_bin[i*S:(i+1)*S, j*S:(j+1)*S]
                if patch.size==0: continue
                if np.any(patch): count += 1
        counts.append(count)
    if len(counts) < 2 or any(c==0 for c in counts): return float('nan')
    coeffs = np.polyfit(np.log(sizes[:len(counts)]), np.log(counts), 1)
    return -coeffs[0]

def vessel_width_metrics(mask):
    mask_bin = (mask>0).astype(np.uint8)
    if mask_bin.sum()==0:
        return {"mean_width":float('nan'), "median_width":float('nan'), "area":0}
    sk = skeletonize(mask_bin>0).astype(np.uint8)
    dist = ndi.distance_transform_edt(mask_bin)
    yx = np.where(sk>0)
    if len(yx[0])==0:
        widths = 2*dist[mask_bin>0]
        if len(widths)==0:
            return {"mean_width":float('nan'), "median_width":float('nan'), "area":int(mask_bin.sum())}
        return {"mean_width":float(np.mean(widths)), "median_width":float(np.median(widths)), "area":int(mask_bin.sum())}
    radii = dist[yx]
    widths = 2*radii
    widths = widths[~np.isnan(widths)]
    if len(widths)==0:
        return {"mean_width":float('nan'), "median_width":float('nan'), "area":int(mask_bin.sum())}
    return {"mean_width":float(np.mean(widths)), "median_width":float(np.median(widths)), "area":int(mask_bin.sum())}

def vessel_tortuosity(mask):
    sk = skeletonize((mask>0).astype(np.uint8)).astype(np.uint8)
    if sk.sum()==0: return float('nan')
    lab = label(sk)
    ratios=[]
    for region in regionprops(lab):
        pts = region.coords
        if pts.shape[0] < 10: continue
        path_length = pts.shape[0]
        dists = np.sum((pts[:,None,:] - pts[None,:,:])**2, axis=-1)
        max_idx = np.unravel_index(np.argmax(dists), dists.shape)
        euclid = np.sqrt(dists[max_idx])
        if euclid < 1e-3: continue
        ratios.append(path_length / euclid)
    if len(ratios)==0: return float('nan')
    return float(np.mean(ratios))

# ---------- extract features ----------
mask_files = sorted([f for f in os.listdir(MASK_DIR) if f.lower().endswith((".png",".jpg",".jpeg"))])
if len(mask_files) == 0:
    raise RuntimeError("No mask images found in MASK_DIR: " + MASK_DIR)
print(f"Found {len(mask_files)} mask files (example): {mask_files[:6]}")

rows = []
start = time.time()
for fname in tqdm(mask_files, desc="Extract features"):
    p = os.path.join(MASK_DIR, fname)
    m = cv2.imread(p, cv2.IMREAD_GRAYSCALE)
    if m is None:
        print("Warning: unreadable mask", fname); continue
    # binarize robustly
    _, bm = cv2.threshold(m, 127, 255, cv2.THRESH_BINARY)
    bin_img = (bm>0).astype(np.bool_)

    # morphological opening (if enabled)
    if MORPH_OPEN_RADIUS and MORPH_OPEN_RADIUS > 0:
        try:
            selem = disk(MORPH_OPEN_RADIUS)
            bin_img = opening(bin_img, selem)
        except Exception:
            k = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (MORPH_OPEN_RADIUS*2+1, MORPH_OPEN_RADIUS*2+1))
            bin_img = cv2.morphologyEx(bm, cv2.MORPH_OPEN, k) > 0

    # remove small objects
    if MIN_OBJECT_SIZE and MIN_OBJECT_SIZE > 0:
        try:
            bin_img = remove_small_objects(bin_img.astype(bool), min_size=MIN_OBJECT_SIZE)
            bin_img = bin_img.astype(np.uint8)
        except Exception:
            nb, lbls, stats, centroids = cv2.connectedComponentsWithStats((bin_img>0).astype(np.uint8), connectivity=8)
            cleaned = np.zeros_like(bin_img, dtype=np.uint8)
            for i in range(1, nb):
                area = stats[i, cv2.CC_STAT_AREA]
                if area >= MIN_OBJECT_SIZE:
                    cleaned[lbls == i] = 1
            bin_img = cleaned

    bin_img = (bin_img > 0).astype(np.uint8)
    fd = fractal_box_count(bin_img)
    vw = vessel_width_metrics(bin_img)
    tort = vessel_tortuosity(bin_img)
    area_frac = float(bin_img.sum()) / (bin_img.shape[0] * bin_img.shape[1])
    rows.append({
        "filename": fname,
        "fractal_dim": float(fd) if not math.isnan(fd) else float('nan'),
        "mean_width": vw["mean_width"],
        "median_width": vw["median_width"],
        "vessel_area": int(vw["area"]),
        "area_frac": area_frac,
        "tortuosity": tort
    })

df_feat = pd.DataFrame(rows)
df_feat.to_csv(OUT_FEATURES, index=False)
print("Saved UNet-based features to:", OUT_FEATURES, " Time elapsed: {:.1f}s".format(time.time()-start))
print(df_feat.head())

# ---------- load labels (or create dummy if missing) ----------
if LABELS_CSV and os.path.exists(LABELS_CSV):
    labels_df = pd.read_csv(LABELS_CSV)
    # ensure columns present
    if 'filename' not in labels_df.columns or 'label' not in labels_df.columns:
        # fallback: assume first col is id/filename and last col is label
        labels_df = labels_df.iloc[:, [0, -1]].rename(columns={labels_df.columns[0]:'filename', labels_df.columns[-1]:'label'})
else:
    alt = "/content/entries_100_to_163.csv"
    if os.path.exists(alt):
        labels_df = pd.read_csv(alt).rename(columns={alt.split('/')[-1]:'filename'})
        labels_df['label'] = 0
    else:
        print("No labels file found; creating dummy random labels (NOT clinically meaningful).")
        labels_df = pd.DataFrame({"filename": df_feat['filename'], "label": np.random.randint(0,2,size=len(df_feat))})

labels_df['filename'] = labels_df['filename'].astype(str).apply(lambda x: os.path.basename(x))
df = df_feat.merge(labels_df, on="filename", how="left")
if df['label'].isnull().any():
    missing = df[df['label'].isnull()]['filename'].tolist()
    print("Warning: these feature files have no label and will be dropped:", missing)
    df = df.dropna(subset=['label'])
df['label'] = df['label'].astype(int)

# ---------- prepare data ----------
feat_cols = [c for c in df.columns if c not in ('filename','label')]
X = df[feat_cols].astype(float).fillna(0).values
y = df['label'].values
print("Training samples:", X.shape[0], "Feature columns:", feat_cols)
if X.shape[0] == 0:
    raise RuntimeError("No labeled samples available after merge; cannot train.")

# standardize features (keeps df intact)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ---------- stratified CV to estimate performance ----------
unique_labels = set(y)
if len(unique_labels) < 2:
    raise RuntimeError("Only one class present in labels — cannot train a classifier. Provide labels with both classes.")
skf = StratifiedKFold(n_splits=min(N_FOLDS, max(2, len(y))), shuffle=True, random_state=RANDOM_STATE)
cv_aucs = []
fold = 0
for train_idx, val_idx in skf.split(X_scaled, y):
    fold += 1
    Xtr, Xv = X_scaled[train_idx], X_scaled[val_idx]
    ytr, yv = y[train_idx], y[val_idx]
    # handle imbalance: scale_pos_weight = (#neg / #pos)
    neg = np.sum(ytr == 0); pos = np.sum(ytr == 1)
    scale_pos_weight = (neg / pos) if pos > 0 else 1.0
    dtrain = xgb.DMatrix(Xtr, label=ytr, feature_names=feat_cols)
    dval = xgb.DMatrix(Xv, label=yv, feature_names=feat_cols)
    params = {"objective":"binary:logistic", "eval_metric":"auc", "eta":0.05, "max_depth":4, "seed":RANDOM_STATE,
              "scale_pos_weight": scale_pos_weight}
    bst = xgb.train(params, dtrain, num_boost_round=500, evals=[(dval,"val")], early_stopping_rounds=25, verbose_eval=False)
    pval = bst.predict(dval)
    try:
        auc = roc_auc_score(yv, pval)
    except Exception:
        auc = float('nan')
    cv_aucs.append(auc)
    print(f"Fold {fold} AUC: {auc:.4f}")

print("CV AUCs:", cv_aucs, "Mean AUC:", np.nanmean(cv_aucs))

# ---------- train final model on all data (using scale_pos_weight) ----------
neg_all = np.sum(y == 0); pos_all = np.sum(y == 1)
scale_pos_weight_all = (neg_all/pos_all) if pos_all>0 else 1.0
dall = xgb.DMatrix(X_scaled, label=y, feature_names=feat_cols)
params = {"objective":"binary:logistic", "eval_metric":"auc", "eta":0.05, "max_depth":4, "seed":RANDOM_STATE,
          "scale_pos_weight": scale_pos_weight_all}
bst_final = xgb.train(params, dall, num_boost_round=500, verbose_eval=False)
bst_final.save_model(OUT_MODEL)
print("Saved final XGBoost model to", OUT_MODEL)

# ---------- predict on all samples and save outputs ----------
probs_all = bst_final.predict(dall)
preds_all = (probs_all >= 0.5).astype(int)
df_out = df.copy()
df_out['pred_prob'] = probs_all
df_out['pred_label'] = preds_all
df_out.to_csv(OUT_FEATURES_PRED, index=False)
print("Saved features+predictions to:", OUT_FEATURES_PRED)

# ---------- evaluation on labeled rows ----------
y_true = df_out['label'].astype(int).values
y_prob = df_out['pred_prob'].values
y_pred = df_out['pred_label'].values
auc_final = roc_auc_score(y_true, y_prob) if len(set(y_true))>1 else float('nan')
acc_final = accuracy_score(y_true, y_pred)
crep = classification_report(y_true, y_pred, digits=4)
cm = confusion_matrix(y_true, y_pred)
print("Final Eval rows:", len(y_true), "AUC:", auc_final, "Acc:", acc_final)
print("Confusion matrix:\n", cm)
print("Classification report:\n", crep)
with open(OUT_REPORT, "w") as f:
    f.write(f"CV AUCs: {cv_aucs}\nMean CV AUC: {np.nanmean(cv_aucs)}\nFinal AUC: {auc_final}\nAccuracy: {acc_final}\n\nConfusion matrix:\n{cm}\n\nClassification report:\n{crep}\n")
print("Saved classification report to", OUT_REPORT)

# ---------- feature importance ----------
fi = bst_final.get_score(importance_type='gain')
fi_df = pd.DataFrame([{"feature":k, "gain":v} for k,v in fi.items()])
# ensure all features present
for c in feat_cols:
    if c not in fi_df['feature'].values:
        fi_df = pd.concat([fi_df, pd.DataFrame([{"feature":c,"gain":0.0}])], ignore_index=True)
fi_df = fi_df.groupby('feature', as_index=False).sum().sort_values('gain', ascending=False)
fi_df.to_csv(OUT_FI, index=False)
print("Saved feature importances to:", OUT_FI)
print(fi_df)

print("Done. Outputs:\n - features: {}\n - features+preds: {}\n - model: {}\n - feature importance: {}\n - report: {}".format(OUT_FEATURES, OUT_FEATURES_PRED, OUT_MODEL, OUT_FI, OUT_REPORT))
```

```python
# Fixed EfficientNet training cell (avoids pandas reindex error when creating numeric column)
import os, random, math, cv2, numpy as np, pandas as pd, time, re
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import models, transforms
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, accuracy_score
import matplotlib.pyplot as plt

# ---------- Config ----------
IMAGES_DIR = "/content/drive/MyDrive/rfmid/train_sample"  # update if needed
LABELS_PATH = "/content/labels_100_163.csv"            # your prepared labels
UNET_MASK_DIR = "/content/unet_refined_masks"
USE_MASK_OVERLAY = True
IMG_SIZE_CLS = 300 # <-- NOTE: EfficientNet-B3 uses 300x300 images
BATCH_SIZE = 8
EPOCHS = 12
LR = 2e-4
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
SEED = 42
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
print("Device:", DEVICE)

# ---------- helpers ----------
def apply_mask_overlay_rgb(img_rgb, mask_gray, alpha=0.35):
    mask_resized = cv2.resize(mask_gray, (img_rgb.shape[1], img_rgb.shape[0]), interpolation=cv2.INTER_NEAREST)
    red_mask = np.zeros_like(img_rgb)
    red_mask[..., 0] = mask_resized
    blended = (img_rgb.astype(np.float32) * (1.0 - alpha) + red_mask.astype(np.float32) * alpha)
    return np.clip(blended, 0, 255).astype(np.uint8)

class FundusDataset(Dataset):
    def __init__(self, df, images_dir, mask_dir=None, use_mask_overlay=False, img_preproc_fn=None, transform=None):
        self.df = df.reset_index(drop=True)
        self.images_dir = images_dir
        self.mask_dir = mask_dir
        self.use_mask_overlay = use_mask_overlay and (mask_dir is not None)
        self.img_preproc_fn = img_preproc_fn
        self.transform = transform
    def __len__(self): return len(self.df)
    def __getitem__(self, idx):
        row = self.df.loc[idx]
        fname = str(row['filename'])
        label = float(row['label'])
        img_path = os.path.join(self.images_dir, fname)
        bgr = cv2.imread(img_path)
        if bgr is None:
            raise FileNotFoundError("Image not found: " + img_path)
        if self.img_preproc_fn is not None:
            # Use the correct image size for B3
            img_rgb = self.img_preproc_fn(bgr, out_size=(IMG_SIZE_CLS, IMG_SIZE_CLS), apply_clahe=True, blur_ksize=(5,5))
        else:
            img_rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)
            img_rgb = cv2.resize(img_rgb, (IMG_SIZE_CLS, IMG_SIZE_CLS))
        if self.use_mask_overlay:
            mask_path = os.path.join(self.mask_dir, fname)
            if os.path.exists(mask_path):
                mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
                img_rgb = apply_mask_overlay_rgb(img_rgb, mask, alpha=0.35)
        img_pil = Image.fromarray(img_rgb)
        if self.transform:
            img_tensor = self.transform(img_pil)
        else:
            img_tensor = transforms.ToTensor()(img_pil)
        return img_tensor, torch.tensor(label, dtype=torch.float32)

# ---------- Load & normalize labels robustly ----------
labels_df = pd.read_csv(LABELS_PATH)
print("Raw label columns:", list(labels_df.columns))
if set(['filename','label']).issubset(set(labels_df.columns)):
    working = labels_df[['filename','label']].copy()
else:
    file_col, label_col = None, None
    for c in labels_df.columns:
        lc = c.lower()
        if any(k in lc for k in ("file","name","id","image")) and file_col is None:
            file_col = c
        if any(k in lc for k in ("label","class","target","risk")) and label_col is None:
            label_col = c
    if file_col is None: file_col = labels_df.columns[0]
    if label_col is None: label_col = labels_df.columns[-1]
    working = labels_df[[file_col, label_col]].rename(columns={file_col:'filename', label_col:'label'}).copy()

# create a set of actual image filenames for robust mapping
image_names = set([f for f in os.listdir(IMAGES_DIR) if f.lower().endswith(('.png','.jpg','.jpeg'))])
print("Images available in folder:", len(image_names))

def normalize_label_name_scalar(x):
    s = str(x).strip()
    bn = os.path.basename(s)
    if bn in image_names:
        return bn
    m = re.search(r'(\d{2,4})', s)
    if m:
        num = m.group(1)
        candidates = [f"Copy of {num}.png", f"{num}.png", f"{num}.PNG"]
        for cand in candidates:
            if cand in image_names:
                return cand
        for cand in image_names:
            if re.search(r'\b' + re.escape(num) + r'\b', cand):
                return cand
    for cand in image_names:
        if s.lower() in cand.lower() or cand.lower() in s.lower():
            return cand
    return None

working['filename_norm'] = working['filename'].apply(normalize_label_name_scalar)
missing_map = int(working['filename_norm'].isna().sum())
if missing_map > 0:
    print(f"Warning: {missing_map} label rows could not be mapped to image filenames automatically.")

mapped = working.dropna(subset=['filename_norm']).copy()

# --- FIX: Drop the original 'filename' column before renaming 'filename_norm' ---
# This prevents a name collision where mapped['filename'] becomes a DataFrame
if 'filename' in mapped.columns:
    mapped = mapped.drop(columns=['filename'])

# Now, rename filename_norm to filename and select the two columns we need
mapped = mapped.rename(columns={'filename_norm':'filename'})[['filename','label']].copy()
# --- END FIX ---

mapped['label'] = mapped['label'].astype(int)
print("Mapped label rows (pre-filter):", len(mapped))

# ---------- Extract numeric ID from filename safely (compute list then assign) ----------
def extract_numeric_from_fname(fname):
    if not isinstance(fname, str):
        return None
    m = re.search(r'(\d{2,4})', fname)
    if not m:
        return None
    try:
        return int(m.group(1))
    except:
        return None

# This line will now work, as mapped['filename'] is correctly a Series
nums = [extract_numeric_from_fname(x) for x in mapped['filename'].tolist()]
# assign as plain list to avoid pandas alignment/reindex issues
mapped['num'] = nums

# filter to 100..163
mapped_subset = mapped[(mapped['num'].notnull()) & (mapped['num'] >= 100) & (mapped['num'] <= 163)].drop(columns=['num']).reset_index(drop=True)
print("Rows after filtering to 100..163:", len(mapped_subset))
if len(mapped_subset) == 0:
    raise RuntimeError("No labels map to images 100..163. Check LABELS_PATH contents and image filenames.")
labels_df = mapped_subset.copy()

# ---------- consistency checks ----------
print("Labels used:", len(labels_df))
available_100_163 = sum(1 for name in image_names if (re.search(r'(\d{2,4})', name) and 100 <= int(re.search(r'(\d{2,4})', name).group(1)) <= 163))
print("Images present in folder with numeric 100..163:", available_100_163)
if available_100_163 != len(labels_df):
    print("Warning: mismatch between mapped labels and image files count. Check filenames and mapping.")

# ---------- Transforms ----------
# Note: Using IMG_SIZE_CLS (now 300) for transforms
train_transforms = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.Resize((IMG_SIZE_CLS, IMG_SIZE_CLS)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
])
val_transforms = transforms.Compose([
    transforms.Resize((IMG_SIZE_CLS, IMG_SIZE_CLS)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485,0.456,0.406], std=[0.229,0.224,0.225])
])

# ---------- Train/Val split ----------
if len(labels_df) < 4:
    raise RuntimeError("Too few mapped samples for training. Need at least 4.")
train_df, val_df = train_test_split(labels_df, test_size=0.2, stratify=labels_df['label'] if len(set(labels_df['label']))>1 else None, random_state=SEED)
train_ds = FundusDataset(train_df, IMAGES_DIR, mask_dir=UNET_MASK_DIR, use_mask_overlay=USE_MASK_OVERLAY, img_preproc_fn=preprocess_raw_image_cv2, transform=train_transforms)
val_ds   = FundusDataset(val_df, IMAGES_DIR, mask_dir=UNET_MASK_DIR, use_mask_overlay=USE_MASK_OVERLAY, img_preproc_fn=preprocess_raw_image_cv2, transform=val_transforms)
train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, shuffle=True, num_workers=2, pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=2, pin_memory=True)
print("Train/Val sizes:", len(train_ds), len(val_ds))

# ---------- Model ----------
# --- FIX 1: Changed to efficientnet_b3 ---
model = models.efficientnet_b3(pretrained=True)
in_feat = model.classifier[1].in_features
model.classifier = nn.Sequential(nn.Dropout(0.4), nn.Linear(in_feat, 1))
model = model.to(DEVICE)
print("Model ready. Params (M):", sum(p.numel() for p in model.parameters())/1e6)

criterion = nn.BCEWithLogitsLoss()
optimizer = optim.AdamW(model.parameters(), lr=LR, weight_decay=1e-5)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, factor=0.5, patience=3)

# ---------- Training loop ----------
BEST_PATH = "/content/efficientnet_b3_best.pth" # Changed path name
best_auc = 0.0
for epoch in range(1, EPOCHS+1):
    model.train()
    running_loss = 0.0
    t0 = time.time()
    for imgs, labels in train_loader:
        imgs = imgs.to(DEVICE); labels = labels.to(DEVICE)
        optimizer.zero_grad()
        logits = model(imgs).squeeze(1)
        loss = criterion(logits, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * imgs.size(0)
    train_loss = running_loss / max(1, len(train_loader.dataset))
    # validation
    model.eval()
    y_true=[]; y_prob=[]
    with torch.no_grad():
        for imgs, labels in val_loader:
            imgs = imgs.to(DEVICE)
            logits = model(imgs).squeeze(1)
            probs = torch.sigmoid(logits).cpu().numpy()
            y_prob.extend(probs.tolist())
            y_true.extend(labels.numpy().tolist())
    if len(set(y_true)) > 1:
        auc = roc_auc_score(y_true, y_prob)
    else:
        auc = float('nan')
    preds = [1 if p>=0.5 else 0 for p in y_prob]
    acc = accuracy_score(y_true, preds) if len(y_true)>0 else float('nan')
    print(f"Epoch {epoch}/{EPOCHS} \ttrain_loss={train_loss:.4f} \tval_auc={auc:.4f} \tval_acc={acc:.4f} \ttime={time.time()-t0:.1f}s")
    if not math.isnan(auc) and auc > best_auc:
        best_auc = auc
        # --- FIX 2: Corrected typo state_Ddict to state_dict ---
        torch.save(model.state_dict(), BEST_PATH)
        print("Saved best model ->", BEST_PATH)
    scheduler.step(0.0 if math.isnan(auc) else auc)

print("Training complete. Best val AUC:", best_auc)
if os.path.exists(BEST_PATH):
    print("Best model saved at", BEST_PATH)

# ---------- Quick inference preview ----------
model.eval()
if len(val_df) > 0:
    samples = val_df.sample(min(4, len(val_df)), random_state=SEED).reset_index(drop=True)
    fig=plt.figure(figsize=(12,6))
    for i, row in samples.iterrows():
        fname = row['filename']; lbl = int(row['label'])
        p = os.path.join(IMAGES_DIR, fname)
        bgr = cv2.imread(p)
        # Use correct image size for B3
        img_rgb = preprocess_raw_image_cv2(bgr, out_size=(IMG_SIZE_CLS, IMG_SIZE_CLS), apply_clahe=True, blur_ksize=(5,5))
        if USE_MASK_OVERLAY:
            maskp = os.path.join(UNET_MASK_DIR, fname)
            if os.path.exists(maskp):
                mask = cv2.imread(maskp, cv2.IMREAD_GRAYSCALE)
                img_rgb = apply_mask_overlay_rgb(img_rgb, mask, alpha=0.35)
        inp = Image.fromarray(img_rgb)
        t = val_transforms(inp).unsqueeze(0).to(DEVICE)
        with torch.no_grad():
            prob = float(torch.sigmoid(model(t).squeeze(1)).cpu().numpy())
        ax = fig.add_subplot(1, min(4, len(samples)), i+1)
        ax.imshow(img_rgb); ax.set_title(f"label={lbl}\nprob={prob:.3f}"); ax.axis('off')
    plt.tight_layout(); plt.show()
else:
    print("No validation samples to preview.")
```

```python
# Only run if you want an image-based model trained on the generated masks and you have labels.
import torchvision.models as models
import torch.nn as nn, torch, torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import albumentations as A
from albumentations.pytorch import ToTensorV2
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, accuracy_score
from PIL import Image
import os, numpy as np, pandas as pd

labels_df = pd.read_csv(labels_path)
# ensure filenames are present in mask folder
labels_df = labels_df[labels_df['filename'].isin(sorted(os.listdir(OUT_MASK_DIR)))].reset_index(drop=True)
if len(labels_df) < 10:
    print("Too few labeled samples to train ResNet (need >=10). Found:", len(labels_df))
else:
    CLS_IMG = 224
    tr, vl = train_test_split(labels_df, test_size=0.2, stratify=labels_df['label'], random_state=SEED)
    class MaskDataset(Dataset):
        def __init__(self, df, masks_dir, transform=None):
            self.df = df.reset_index(drop=True); self.masks_dir = masks_dir; self.transform = transform
        def __len__(self): return len(self.df)
        def __getitem__(self, idx):
            fname = self.df.loc[idx,'filename']; lbl = int(self.df.loc[idx,'label'])
            p = os.path.join(self.masks_dir, fname)
            im = np.array(Image.open(p).convert("RGB"))
            if self.transform: im = self.transform(image=im)['image']
            return im, torch.tensor(lbl, dtype=torch.float32)
    train_t = A.Compose([A.Resize(CLS_IMG,CLS_IMG), A.RandomRotate90(), A.Flip(), A.Normalize(mean=(0.485,0.456,0.406), std=(0.229,0.224,0.225)), ToTensorV2()])
    val_t   = A.Compose([A.Resize(CLS_IMG,CLS_IMG), A.Normalize(mean=(0.485,0.456,0.406), std=(0.229,0.224,0.225)), ToTensorV2()])
    train_ds = MaskDataset(tr, OUT_MASK_DIR, transform=train_t); val_ds = MaskDataset(vl, OUT_MASK_DIR, transform=val_t)
    train_loader = DataLoader(train_ds, batch_size=8, shuffle=True, num_workers=2, pin_memory=True)
    val_loader = DataLoader(val_ds, batch_size=8, shuffle=False, num_workers=2, pin_memory=True)

    device = DEVICE
    resnet = models.resnet50(pretrained=True)
    in_features = resnet.fc.in_features
    resnet.fc = nn.Linear(in_features, 1)
    model = resnet.to(device)
    criterion = nn.BCEWithLogitsLoss()
    opt = optim.AdamW(model.parameters(), lr=2e-4, weight_decay=1e-5)

    best_auc = 0.0
    for epoch in range(1, 8):
        model.train()
        for imgs, labels in train_loader:
            imgs = imgs.to(device); labels = labels.to(device)
            opt.zero_grad(); logits = model(imgs).squeeze(1); loss = criterion(logits, labels); loss.backward(); opt.step()
        # eval
        model.eval()
        ys, ps = [], []
        with torch.no_grad():
            for imgs, labels in val_loader:
                imgs = imgs.to(device)
                logits = model(imgs).squeeze(1)
                probs = torch.sigmoid(logits).cpu().numpy()
                ys.extend(labels.numpy().tolist()); ps.extend(probs.tolist())
        auc = roc_auc_score(ys, ps) if len(set(ys))>1 else float('nan')
        preds = [1 if p>=0.5 else 0 for p in ps]; acc = accuracy_score(ys, preds)
        print(f"Epoch {epoch} val_auc={auc:.4f} val_acc={acc:.4f}")
        if not np.isnan(auc) and auc > best_auc:
            best_auc = auc; torch.save(model.state_dict(), "/content/resnet_masks_best.pth"); print("Saved best resnet")
    print("Finished ResNet training. Best AUC:", best_auc)
```

```python
import statsmodels.api as sm
X = df[feat_cols].astype(float)
Xz = (X - X.mean())/X.std()            # standardized
Xz = sm.add_constant(Xz)
y = df['label']
logit = sm.Logit(y, Xz).fit(disp=False)
print(logit.summary())

# Odds ratios (+95% CI)
params = logit.params
conf = logit.conf_int()
odds = np.exp(params)
odds_ci = np.exp(conf)
odds_df = pd.concat([odds, odds_ci], axis=1).rename(columns={0:'2.5%',1:'97.5%'})
odds_df.columns = ['odds_ratio','2.5%','97.5%']
print("\nOdds ratios (std features):\n", odds_df)
```

```python
import xgboost as xgb
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import roc_auc_score, accuracy_score, classification_report

X = df[feat_cols].fillna(0).values
y = df['label'].values

kf = StratifiedKFold(n_splits=min(5, len(y)), shuffle=True, random_state=42)
aucs=[]
preds = np.zeros(len(y))
for train_idx, test_idx in kf.split(X,y):
    Xtr, Xte = X[train_idx], X[test_idx]
    ytr, yte = y[train_idx], y[test_idx]
    # handle imbalance
    neg = (ytr==0).sum(); pos = (ytr==1).sum()
    spw = (neg/pos) if pos>0 else 1.0
    dtrain = xgb.DMatrix(Xtr, label=ytr)
    dtest  = xgb.DMatrix(Xte, label=yte)
    params = {"objective":"binary:logistic","eval_metric":"auc","eta":0.05,"max_depth":4,"seed":42,"scale_pos_weight":spw}
    bst = xgb.train(params, dtrain, num_boost_round=200, evals=[(dtest,'val')], early_stopping_rounds=20, verbose_eval=False)
    p = bst.predict(dtest)
    preds[test_idx] = p
    try:
        auc = roc_auc_score(yte, p)
    except:
        auc = float('nan')
    aucs.append(auc)
    print("Fold AUC:", auc)
print("CV AUCs:", aucs, "Mean AUC:", np.nanmean(aucs))

# final metrics using CV preds
y_pred = (preds >= 0.5).astype(int)
print("Overall CV Acc:", accuracy_score(y, y_pred))
print(classification_report(y, y_pred, digits=4))
```
